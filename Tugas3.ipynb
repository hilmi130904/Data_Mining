```python
df = df.drop_duplicates(subset='NAMA', keep='first')
df
```

```python
#mencari baris yang mengandung missing values
baris_dengan_data_hilang=df[df.isnull().any(axis=1)]
baris_dengan_data_hilang
```

```python
#mencari kolom yang mengandung missing values
kolom_dengan_data_hilang=df.columns[df.isnull().any()]
kolom_dengan_data_hilang
```
